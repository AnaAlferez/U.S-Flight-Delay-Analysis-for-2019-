```python
#Cleaning Data
#Replace NA's with E category for CANCELLATION_CODE
data['CANCELLATION_CODE']=data['CANCELLATION_CODE'].fillna('E')

#Replace NA's with 999999 for all other columns
data = data.fillna(999999)

#Converting to numeric value for the required columns  
lst = ['Flights', 'DEP_DELAY', 'DEP_DEL15', 'TAXI_OUT', 'TAXI_IN',
       'ARR_DELAY', 'CANCELLED', 'DIVERTED', 'AIR_TIME',
       'DISTANCE', 'DISTANCE_GROUP', 'CARRIER_DELAY',
       'SECURITY_DELAY', 'LATE_AIRCRAFT_DELAY', 'DIV_AIRPORT_LANDINGS',
       'DIV_ARR_DELAY']
for i in lst:
    data[i]=data[i].astype(int)
data
```

```python
#Evaluating DEPARTURE DELAY OF AT LEAST 15 MINUTES WITH logistic regression model.
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split

datacor=datacor.query('AIR_TIME != 999999')
datacor=datacor.query('TAXI_OUT != 999999')
datacor=datacor.query('DISTANCE_GROUP != 999999')
datacor=datacor.query('DEP_DEL15 != 999999')

x_data =datacor[['AIR_TIME','TAXI_OUT','DISTANCE_GROUP']]
y_data =datacor['DEP_DEL15']
x_training_data, x_test_data, y_training_data, y_test_data = train_test_split(x_data, y_data, test_size = 0.3)
model = LogisticRegression()
model.fit(x_training_data, y_training_data)
predictions = model.predict(x_test_data)
print(classification_report(y_test_data, predictions))
```

```python
#Evaluation of Logistic Regression
import statsmodels.formula.api as smf
import seaborn as sns
import statsmodels.formula.api as smf
model= smf.logit(formula="DEP_DEL15~ AIR_TIME + TAXI_OUT + DISTANCE_GROUP", data= datacor).fit()
model.summary()

#Interpretation:
#For every one unit increase in assign_days, the log odds of a lead being assigned to a sales representative increase by 0.0052
#And ,for every one unit increase in age, the log odds of a lead being assigned to a sales representative increase by 0.0060.
```

```python
#Converting the coefficients of the logistic regression model into odd ratios.Getting the exponent of the coefficient value.
model_odds = pd.DataFrame(np.exp(model.params), columns= ['OR'])
model_odds['z-value']= model.pvalues
model_odds[['2.5%', '97.5%']] = np.exp(model.conf_int())
model_odds

#All the IV's have a significant effect on the log odds of being admitted as indicated by the z values < 0.05.
#assign_days has a z-value 2.16x10-12 and age 1.43x10-81 which are less than 0.05, that is why it is admited.

#Once the log coefficients were converted to odds ratios OR, we say:
#For every one unit increase in assign_days, the log odds of a lead being assigned to a sales representative increases by a factor of  1.00519.
#And ,for every one unit increase in age, the log odds of a lead being assigned to a sales representative increases by a factor of  1.0060.
```

```python
#RANDOM FOREST
from sklearn.model_selection import train_test_split
dataRF=data.iloc[:100000]
# Es buena práctica definir una lista con los nombres de las variables explicativas y la variable objetivo a predecir
predictores = ['TAXI_OUT','TAXI_IN','ARR_DELAY','AIR_TIME','DISTANCE','DISTANCE_GROUP']

objetivo = ['DEP_DELAY']

# División train-test!
X_entrenamiento, X_prueba, Y_entrenamiento, Y_prueba= train_test_split(
    dataRF[predictores], # aquí se introduce el conjunto de variables predictoras
    dataRF[objetivo], # aquí se introduce el objetivo a predecir
    test_size = 0.2, # indicamos que queremos entrenar sobre el 80% de datos y evaluar sobre el 20% restante
    random_state = 1234# ES como la misma semilla, para tener los mismos resultados
)
```

```python
# Verificamos que las divisiones se hicieron correctamente, observando las dimensiones:
print(X_entrenamiento.shape)
print(X_prueba.shape)
print(Y_entrenamiento.shape)
print(Y_prueba.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor

## Hiperparámetros (configuraciones 'padre')
# Es muy buena práctica definir las configuraciones del modelo en la parte superior de cada celda,
# lo cual nos permite realizar experimentos de manera ordenada.
n_arboles = 50
profundidad_maxima = 12
casos_minimos_nodo = 50
cantidad_maxima_variables = 'auto' # puede ser un número, 'sqrt' (raiz del numero de variables) o 'auto' (todas las variables)

# Primero se crea el objeto 'modelo'. Este contiene la estructura general del algoritmo y es posible acceder a su
# contenido invocando su nombre
RF = RandomForestRegressor(
    n_estimators = n_arboles,
    max_depth = profundidad_maxima,
    min_samples_leaf = casos_minimos_nodo,
    max_features = cantidad_maxima_variables,
    random_state = 1234 # establecemos una semilla aleatoria para reproducir resultados
)
```

```python
# Ajustando con el método ".fit()"

RF.fit(
    X = X_entrenamiento,
    y = Y_entrenamiento.values.ravel()
)
```

```python
# Utilizamos el modelo ya entrenado:
predicciones_entrenamiento = RF.predict(X_entrenamiento) # predicción sobre los datos que el modelo ya vió
predicciones_prueba = RF.predict(X_prueba) # predicción sobre los datos que el modelo no vió

# Revisamos las filas de las predicciones y se observa que calza con las dimensiones de X_entrenamiento:
print(X_entrenamiento.shape)
print(predicciones_entrenamiento.shape) # única columna

# Revisamos las filas de las predicciones y se observa que calza con las dimensiones de X_prueba:
print(X_prueba.shape)
print(predicciones_prueba.shape) # única columna
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Primero evaluamos el ajuste sobre el conjunto de entrenamiento.
# La lógica es la siguiente: Si el modelo no puede predecir bien lo que ya observó, entonces
# no tiene sentido que se utilce para predecir información nueva.
print(
    "Métricas para el conjunto de entrenamiento: \n R2: {} \n MAE: {} \n MSE: {}"\
    .format(
        r2_score(
            y_true = Y_entrenamiento, # los datos REALES
            y_pred = predicciones_entrenamiento # los datos estimados por el modelo
        ),
        mean_absolute_error(
            y_true = Y_entrenamiento,
            y_pred = predicciones_entrenamiento
        ),
        mean_squared_error(
            y_true = Y_entrenamiento,
            y_pred = predicciones_entrenamiento
        )
    )
)
```

```python
# Primero evaluamos el ajuste sobre el conjunto de prueba.

# Los resultados de las métricas sobre este conjunto de prueba son los que determinan si es recomendable
# aceptar o no el modelo
print(
    "Métricas para el conjunto de prueba: \n R2: {} \n MAE: {} \n MSE: {}"\
    .format(
        r2_score(
            y_true = Y_prueba, # los datos REALES
            y_pred = predicciones_prueba # los datos estimados por el modelo
        ),
        mean_absolute_error(
            y_true = Y_prueba,
            y_pred = predicciones_prueba
        ),
        mean_squared_error(
            y_true = Y_prueba,
            y_pred = predicciones_prueba
        )
    )
)
```

```python
# Primero definimos una función para graficar la importancia de variables que el model asigna:
def PlotearImportanciaVariables(modelo, lista_predictores, figsize=(10,10)):
    fig, ax1 = plt.subplots(1, 1, figsize=figsize)
    orden = modelo.feature_importances_.argsort().tolist()
    orden.reverse()
    sns.barplot(modelo.feature_importances_[orden], 
            [lista_predictores[i] for i in orden],
            orient='h', ax=ax1)
    return None
```

```python
# Graficando!
PlotearImportanciaVariables(
    modelo = RF,
    lista_predictores = predictores,
    figsize = (15,15) # podemos ajustar el tamaño de la imagen
)
```

```python
#KNeighborsClassifier
from sklearn import model_selection
from sklearn.neighbors import KNeighborsClassifier

datacor=datacor.iloc[:100000]
x_data =datacor[['ARR_DELAY','AIR_TIME','TAXI_OUT','TAXI_IN']]  #data
y_data =datacor['DEP_DELAY']  #y

X_trainK, X_testK, y_trainK, y_testK = train_test_split(x_data, y_data, test_size = 0.3, random_state = 101)

#Neighbors
neighbors = np.arange(0,25)

#Create empty list that will hold cv scores
cv_scores = []

#Perform 10-fold cross validation on training set for odd values of k:
for k in neighbors:
    k_value = k+1
    knn = KNeighborsClassifier(n_neighbors = k_value, weights='uniform', p=2, metric='euclidean')
    kfold = model_selection.KFold(n_splits=10, random_state=None)
    scores = model_selection.cross_val_score(knn, X_trainK, y_trainK, cv=kfold, scoring='accuracy')
    cv_scores.append(scores.mean()*100)
    print("k=%d %0.2f (+/- %0.2f)" % (k_value, scores.mean()*100, scores.std()*100))

optimal_k = neighbors[cv_scores.index(max(cv_scores))]
print ("The optimal number of neighbors is %d with %0.1f%%" % (optimal_k, cv_scores[optimal_k]))

plt.plot(neighbors, cv_scores)
plt.xlabel('Number of Neighbors K')
plt.ylabel('Train Accuracy')
plt.show()
```
